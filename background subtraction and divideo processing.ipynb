```python
import cv2

video_path = r"Downloads/27260-362770008_large.mp4"
cap = cv2.VideoCapture(video_path)

bg_subtractor = cv2.createBackgroundSubtractorMOG2()


frame_width = 640
frame_height = 480

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
 
    frame = cv2.resize(frame, (frame_width, frame_height))
    
    fg_mask = bg_subtractor.apply(frame)
    
 
    _, fg_mask = cv2.threshold(fg_mask, 250, 255, cv2.THRESH_BINARY)
    
  
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Foreground Mask", fg_mask)

   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


video_path = r"Downloads/27260-362770008_large.mp4"
cap = cv2.VideoCapture(video_path)

alpha = 0.1 
avg_background = None


frame_width = 640
frame_height = 480

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    

    frame = cv2.resize(frame, (frame_width, frame_height))
    

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

  
    if avg_background is None:
        avg_background = np.float32(gray_frame)
    else:
    
    
    avg_background_frame = cv2.convertScaleAbs(avg_background)
    
   
    fg_mask = cv2.absdiff(gray_frame, avg_background_frame)
    
   
    _, fg_mask = cv2.threshold(fg_mask, 50, 255, cv2.THRESH_BINARY)
    
 
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Foreground Mask", fg_mask)

   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

threshold = 0.5 
cap = cv2.VideoCapture(r"Downloads/Screen Recording 2025-03-31 212851.mp4")
frame_count = 0
prev_frame = None


frame_width = 640
frame_height = 480

def calculate_histogram_difference(frame1, frame2):
    hist1 = cv2.calcHist([frame1], [0], None, [256], [0, 256])
    hist2 = cv2.calcHist([frame2], [0], None, [256], [0, 256])
    
    cv2.normalize(hist1, hist1, 0, 1, cv2.NORM_MINMAX)
    cv2.normalize(hist2, hist2, 0, 1, cv2.NORM_MINMAX)
    diff = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)
    
    return diff

while True:
    ret, frame = cap.read()
    if not ret:
        print("End of video or error reading frame.")
        break
    
    frame_count += 1
    
  
    frame = cv2.resize(frame, (frame_width, frame_height))
    
  
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    if prev_frame is not None:

        diff = calculate_histogram_difference(prev_frame, gray)
        
     
        if diff < threshold:
            print(f"Scene Change Detected at frame {frame_count} (Difference: {diff:.2f})")
        
   
        cv2.putText(frame, f"Diff: {diff:.2f}", (10, 30), 
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
   
    prev_frame = gray
    
  
    cv2.imshow("Video Frame", frame)
    
   
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
