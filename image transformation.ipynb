```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Downloads/wallpapers.jpg')
print(f"Image size: {image.shape[0]} x {image.shape[1]} (Height x Width)")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
def translate_image(image, tx, ty):
    rows, cols, channels = image.shape
    translated_image = np.zeros_like(image)
    for i in range(rows):
        for j in range(cols):
            new_x = i + ty
            new_y = j + tx
            if 0 <= new_x < rows and 0 <= new_y < cols:
                translated_image[new_x, new_y] = image[i, j]
    return translated_image


translated_image = translate_image(image, 100, 50)
display_image('Translated Image', translated_image)
```

```python
def rotate_image(image, angle):
    angle = np.deg2rad(angle)
    rows, cols, channels = image.shape
    center_x, center_y = rows // 2, cols // 2
    rotated_image = np.zeros_like(image)
    for i in range(rows):
        for j in range(cols):
            new_x = int((i - center_x) * np.cos(angle) - (j - center_y) * np.sin(angle) + center_x)
            new_y = int((i - center_x) * np.sin(angle) + (j - center_y) * np.cos(angle) + center_y)
            if 0 <= new_x < rows and 0 <= new_y < cols:
                rotated_image[new_x, new_y] = image[i, j]
    return rotated_image

rotated_image = rotate_image(image, 45)
display_image('Rotated Image', rotated_image)
```

```python
def scale_image(image, fx, fy):
    rows, cols, channels = image.shape
    new_rows, new_cols = int(rows * fy), int(cols * fx)
    scaled_image = np.zeros((new_rows, new_cols, channels), dtype=image.dtype)
    for i in range(new_rows):
        for j in range(new_cols):
            orig_x = int(i / fy)
            orig_y = int(j / fx)
            if 0 <= orig_x < rows and 0 <= orig_y < cols:
                scaled_image[i, j] = image[orig_x, orig_y]
    return scaled_image

print(f"Image size: {scaled_image.shape[0]} x {scaled_image.shape[1]} (Height x Width)")
scaled_image = scale_image(image, 10,5)
display_image('Scaled Image', scaled_image)
```

```python
def shear_image_x(image, shear_factor_x):
    rows, cols, channels = image.shape
    sheared_image_x = np.zeros_like(image)
    for i in range(rows):
        for j in range(cols):
            new_x = i
            new_y = j + int(shear_factor_x * i)
            if 0 <= new_x < rows and 0 <= new_y < cols:
                sheared_image_x[new_x, new_y] = image[i, j]
    return sheared_image_x

sheared_image_x = shear_image_x(image, 0.2)
display_image('Sheared Image X', sheared_image_x)
```

```python
def shear_image_y(image, shear_factor_y):
    rows, cols, channels = image.shape
    sheared_image_y = np.zeros_like(image)
    for i in range(rows):
        for j in range(cols):
            new_x = i + int(shear_factor_y * j)
            new_y = j
            if 0 <= new_x < rows and 0 <= new_y < cols:
                sheared_image_y[new_x, new_y] = image[i, j]
    return sheared_image_y

sheared_image_y = shear_image_y(image, 0.3)
display_image('Sheared Image Y', sheared_image_y)
```

```python
def reflect_image(image, axis):
    rows, cols = image.shape[:2]
    reflected_image = np.zeros_like(image)
    
    if axis == 'horizontal':
        for i in range(rows):
            for j in range(cols):
                reflected_image[i, j] = image[i, cols - j - 1]
    elif axis == 'vertical':
        for i in range(rows):
            for j in range(cols):
                reflected_image[i, j] = image[rows - i - 1, j]
    elif axis == 'both':
        for i in range(rows):
            for j in range(cols):
                reflected_image[i, j] = image[rows - i - 1, cols - j - 1]
    
    return reflected_image

reflected_image_horizontal = reflect_image(image, 'horizontal')
reflected_image_vertical = reflect_image(image, 'vertical')
reflected_image_both = reflect_image(image, 'both')

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(reflected_image_horizontal, cv2.COLOR_BGR2RGB))
plt.title('Reflected Image Horizontal')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(reflected_image_vertical, cv2.COLOR_BGR2RGB))
plt.title('Reflected Image Vertical')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(reflected_image_both, cv2.COLOR_BGR2RGB))
plt.title('Reflected Image Both')
plt.axis('off')

plt.show()
```

```python
def display_image(title, img):
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```
