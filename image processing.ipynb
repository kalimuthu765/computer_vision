```python
import numpy as np
import cv2 as cv
image=cv.imread("Downloads/Lenna_(test_image).png")
print(f"Image size: {image.shape[0]} x {image.shape[1]} (Height x Width)")
```

```python
import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)  
plt.imshow(image)
plt.title("Original Image(BGR)")
plt.axis('off')
```

```python
import matplotlib.pyplot as plt
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.subplot(1, 2, 1)  
plt.imshow(image_rgb)
plt.title("Original Image (RGB)")
plt.axis('off')
plt.show()
```

```python
 image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
 plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(image_rgb.ravel(), bins=256, range=(0, 256))
plt.title("Histogram of rbg Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(image_gray.ravel(), bins=256, range=(0, 256), color='gray')
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
hist, bins = np.histogram(image_gray.flatten(), bins=256, range=[0, 256])

cdf = hist.cumsum()

cdf_normalized = cdf * 255 / cdf[-1]

image_equalized = np.interp(image_gray.flatten(), bins[:-1], cdf_normalized)
image_equalized = image_equalized.reshape(image_gray.shape).astype(np.uint8)
```

```python
plt.subplot(1, 2, 2)
plt.imshow(image_equalized, cmap='gray')
plt.title("Equalized Grayscale Image")
plt.axis('off')
    
```

```python
plt.subplot(1, 2, 2)
plt.hist(image_equalized.ravel(), bins=256, range=(0, 256), color='gray')
plt.title("Histogram of Equalized Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
```

```python
newwidth, newheight =600,790
resize_image_rgb=cv.resize(image_rgb ,(newwidth, newheight))
plt.subplot(1, 2, 1)  
plt.imshow(resize_image_rgb)
plt.title("resize image(RGB)")
plt.axis('off')
print(f"Image size: {resize_image_rgb.shape[0]} x {resize_image_rgb.shape[1]} (Height x Width)")
```

 ***manually***

```python
 image_gray_manual = 0.2989 * image_rgb[:, :, 2] + 0.5870 * image_rgb[:, :, 1] + 0.1140 * image[:, :, 0]
```

```python
plt.subplot(1, 2, 2)
plt.imshow(image_gray_manual, cmap='gray')
plt.title("Manually Grayscale Image")
plt.axis('off')
```

```python
newwidth, newheight =800,300
resize_image_gray_manual=cv.resize(image_gray_manual  ,(newwidth, newheight))
plt.subplot(1, 2, 1)  
plt.imshow(resize_image_gray_manual,cmap='gray')
plt.title("resize image(gray)")
plt.axis('off')
print(f"Image size: {resize_image_gray_manual.shape[0]} x {resize_image_gray_manual.shape[1]} (Height x Width)")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(image_gray_manual.ravel(), bins=256, range=(0, 256), color='gray')
plt.title("Histogram of Manually Grayscale Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
histogram = np.zeros(256, dtype=int)
height, width = image_gray_manual.shape
for i in range(height):
   for j in range(width):
        pixel_value = int(image_gray_manual[i, j]) 
        histogram[pixel_value] += 1 

cdf = np.cumsum(histogram)

cdf_normalized = (cdf - cdf.min()) / (cdf.max() - cdf.min()) * 255

image_equalized = np.zeros_like(image_gray_manual)
for i in range(height):
    for j in range(width):
         pixel_value = int(image_gray_manual[i, j]) 
         image_equalized[i, j] = np.round(cdf_normalized[pixel_value]) 
```

```python
plt.subplot(1, 2, 2)
plt.imshow(image_equalized, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.hist(image_equalized.ravel(), bins=256, range=(0, 256), color='gray')
plt.title("Histogram of Equalized Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
```
