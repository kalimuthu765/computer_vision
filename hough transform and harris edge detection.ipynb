```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Downloads/brown-kraft-cube-shaped-box-tuck-top-bottom.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray_image, (5, 5), 0)
edges = cv2.Canny(gray_image, 70, 70)
lines = cv2.HoughLines(edges, 1, np.pi / 180,170 )
if lines is not None:
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
a=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(a)
plt.axis('off')
plt.show()
```

```python
image1 = cv2.imread('Downloads/brown-kraft-cube-shaped-box-tuck-top-bottom.png')
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image1 = np.float32(gray_image1)
blurred1 = cv2.GaussianBlur(gray_image1, (5, 5), 0)
dst1 = cv2.cornerHarris(blurred1, 30, 9, 0.06)
dst1 = cv2.dilate(dst1, None)
image1[dst1 > 0.01 * dst1.max()] = [0, 0, 255]
threshold1 = 0.01 * dst1.max()
for i1 in range(dst1.shape[0]):
    for j1 in range(dst1.shape[1]):
        if dst1[i1, j1] > threshold1:
            cv2.circle(image1, (j1, i1), 10, (0, 0, 255), 6)
b1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(b1)
plt.axis('off')
plt.show()
```
